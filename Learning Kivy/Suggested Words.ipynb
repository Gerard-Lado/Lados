```python


searched_word = input("Input Word: ")
searched_word = searched_word.upper()
print(searched_word)
tested_word_list = []
searched_word_list = []
tested_word = ""

for letter in searched_word:
    searched_word_list.append(letter)

for letter in searched_word:
    tested_word_list.append(letter)


print(tested_word)
word_dict = {"ibuhos": "ye", "na": "naw", "ang": "ha", "beer": "pls", "sa": "kk", "aking": "wo", "lalamunan": "mm"}
word_list = []
suggested_words = []


for key in word_dict:
    key = key.upper()
    word_list.append(key)

print(word_list)

keyboard_approximate = {
    "A": "QWSXZ",
    "B": "VGHN",
    "C": "XDFV",
    "D": "SERFCX",
    "E": "WSDFR",
    "F": "DERTGVC",
    "G": "RTYFHVB",
    "H": "TYUGJN",
    "I": "UJKLO",
    "J": "UIHKNM",
    "K": "IOJLM",
    "L": "OKP",
    "M": "NJKL",
    "N": "BHJM",
    "O": "IKLP",
    "P": "OL",
    "Q": "ASW",
    "R": "EDFGT",
    "S": "QWEADZXC",
    "T": "RFGHY",
    "U": "YHJKI",
    "V": "CFGB",
    "W": "QASDE",
    "X": "ZASDC",
    "Y": "TGHJU",
    "Z": "ASX"
}

vowel_approximate = {
    "A": "EIOU",
    "E": "AIOU",
    "I": "AEOU",
    "O": "AEIU",
    "U": "AEIO"
}

for a in range(len(searched_word_list)):
    for key, val in keyboard_approximate.items():
        if tested_word_list[a] == k:
            for approx_letter in val:
                for g in range(len(searched_word_list)):
                    tested_word_list[g] = searched_word_list[g]
                tested_word_list[a] = approx_letter
                tested_word = ""
                print(tested_word) #Helps me debug, Can be removed
                for i in tested_word_list:
                    tested_word = tested_word + i
                for c in word_list:
                    if tested_word == c:
                        suggested_words.append(c)
                    else:
                        pass
    for key, val in vowel_approximate.items():
        if tested_word_list[a] == key:
            for approx_vowel in val:
                for f in range(len(searched_word_list)):
                    tested_word_list[f] = searched_word_list[f]
                tested_word_list[a] = approx_vowel
                tested_word = ""
                for h in tested_word_list:
                    tested_word = tested_word + h
                print(tested_word) #Helps me debug, Can be removed
                for e in word_list:
                    if tested_word == e:
                        suggested_words.append(e)
                    else:
                        pass
    
print(suggested_words)
```
